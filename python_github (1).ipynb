```python
# Scaling/Normalization
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

df[['Age', 'Calories']] = scaler.fit_transform(df[['Age', 'Calories']])
print("Normalized Data (Age & Calories):")
print(df[['Age', 'Calories']].head())
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
df['Calories_Burned'] = encoder.fit_transform(df['Calories_Burned'])
print("\nEncoded 'Calories_Burned':")
print(df[['Calories_Burned']].head())
```

```python

Q1 = df['Age'].quantile(0.25)
Q3 = df['Age'].quantile(0.75)
IQR = Q3 - Q1
lower_bound_age = Q1 - 1.5 * IQR
upper_bound_age = Q3 + 1.5 * IQR



outliers_age = df[(df['Age'] < lower_bound_age) | (df['Age'] > upper_bound_age)]
print("Outliers detected using IQR (Age):")
print(outliers_age)

# Repeat for 'Calories'
Q1_calories = df['Calories'].quantile(0.25)
Q3_calories = df['Calories'].quantile(0.75)
IQR_calories = Q3_calories - Q1_calories
lower_bound_calories = Q1_calories - 1.5 * IQR_calories
upper_bound_calories = Q3_calories + 1.5 * IQR_calories

outliers_calories = df[(df['Calories'] < lower_bound_calories) | (df['Calories'] > upper_bound_calories)]
print("\nOutliers detected using IQR (Calories):")
print(outliers_calories)
```

```python
from scipy import stats

z_scores_age = stats.zscore(df['Age'])
z_scores_calories = stats.zscore(df['Calories'])


threshold = 3

outliers_zscore_age = df[abs(z_scores_age) > threshold]
outliers_zscore_calories = df[abs(z_scores_calories) > threshold]

print("\nOutliers detected using Z-Score (Age):")
print(outliers_zscore_age)

print("\nOutliers detected using Z-Score (Calories):")
print(outliers_zscore_calories)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
sns.boxplot(data=df['Age'])
plt.title('Boxplot for Age')
plt.show()


plt.figure(figsize=(8, 6))
sns.boxplot(data=df['Calories'])
plt.title('Boxplot for Calories')
plt.show()
```

```python
# Remove outliers using IQR for both 'Age' and 'Calories'
df_no_outliers = df[(df['Age'] >= lower_bound_age) & (df['Age'] <= upper_bound_age) &
                    (df['Calories'] >= lower_bound_calories) & (df['Calories'] <= upper_bound_calories)]

print("\nData after removing outliers:")
print(df_no_outliers)
```

```python
# Capping outliers for Age (replace values outside the bounds)
df['Age'] = df['Age'].clip(lower=lower_bound_age, upper=upper_bound_age)

# Capping outliers for Calories
df['Calories'] = df['Calories'].clip(lower=lower_bound_calories, upper=upper_bound_calories)

print("\nData after capping outliers:")
print(df)
```

```python

df['Age'] = np.log1p(df['Age'])
df['Calories'] = np.log1p(df['Calories'])

print("\nData after log transformation (Age & Calories):")
print(df)
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Select the numerical columns for PCA
numerical_cols = ['Age', 'Calories', 'Weight (kg)', 'Height (m)', 'Fat_Percentage']

# Standardize the data before applying PCA
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df[numerical_cols])

# Apply PCA
pca = PCA(n_components=2)  # Reduce to 2 principal components
pca_result = pca.fit_transform(scaled_data)

# Convert PCA result to DataFrame
pca_df = pd.DataFrame(data=pca_result, columns=['PCA1', 'PCA2'])

# Add the Meal Category to the PCA DataFrame for visualization
pca_df['Calories_Burned'] = df['Calories_Burned']

# Plot the results
plt.figure(figsize=(8, 6))
sns.scatterplot(x='PCA1', y='PCA2', hue='Calories_Burned', data=pca_df, palette='viridis')
plt.title('PCA: Principal Components 1 and 2 by Meal Category')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Calories_Burned')
plt.show()
```
